```python
import re

def text_cleaner(text):
    newString = text.lower()
    newString = re.sub(r'\([^)]*\)', '', newString)
    newString = re.sub('"','', newString)
    newString = ' '.join([contraction_mapping[t] if t in contraction_mapping else t for t in newString.split(" ")])    
    newString = re.sub(r"'s\b","",newString)
    newString = re.sub("[^a-zA-Z]", " ", newString) 
    newString = re.sub('[m]{2,}', 'mm', newString) 
    return newString.strip()
```

```python
data['cleaned_text'] = data['Text'].apply(lambda x: text_cleaner(x))
data['cleaned_summary'] = data['Summary'].apply(lambda x: text_cleaner(x))
data['cleaned_text'] = 'summarize: ' + data['cleaned_text']
data
```

```python
data = data.reset_index(drop=True)
data_for_training = data[['cleaned_text', 'cleaned_summary']]
data_for_training.to_csv('data/dataset.csv', index=False)
data_for_training
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(data_for_training, test_size=0.2)
train.to_csv('data/train.csv', index=False)
test.to_csv('data/test.csv', index=False)
```
